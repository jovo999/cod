```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cod.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
best_player = df.groupby('name')['wins'].sum().reset_index().sort_values(by='wins', ascending=False).head(10)
best_player
```

```python
bad_player = df.groupby('name')['losses'].sum().reset_index().sort_values(by='losses', ascending=False).head(10)
bad_player
```

```python
kd_player = df.groupby('name')['kdRatio'].sum().reset_index().sort_values(by='kdRatio', ascending=False).head(10)
kd_player
```

```python
nkas = df.groupby('name')[['kills' , 'assists' , 'shots']].sum().reset_index().sort_values(by='kills', ascending=False).head(10)
nkas
```

```python
def time_play(time):
    if time <= 4:
      return 'very low'
    elif 4 < time <= 51:
      return 'low'
    elif 51 < time <= 450:
      return 'medum'
    elif 450 < time <= 2000:
      return 'high'
    else:
      return 'very high'

df['active stauts'] = df['timePlayed'].apply(time_play)
```

```python
df.sort_values(['xp'] , ascending=False).head(10)
```

```python
corr_w3 = ['kills' , 'timePlayed' ,'kdRatio' ,'wins' , 'losses']

corr_matrix = df[corr_w3].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix , annot=True)
```

```python
sns.pairplot(df[corr_w3])
```

```python
sns.barplot(x = 'name' , y = 'wins' , data = best_player)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df , x='active stauts')
```
